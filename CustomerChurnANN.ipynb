```python
import pandas as pd


df = pd.read_csv('/Users/facundoginevra/Downloads/BankCustomers.csv')
```

```python
df.info()
```

```python
#'Exited' column is the target/ dependant variable that we are going to predict
```

```python
x = df.drop('Exited', axis=1)
y = df['Exited']
```

```python
x = x.drop(['RowNumber','CustomerId', 'Surname'], axis=1)

```

```python
# We have to convert the categorical (string) variables into numerical

states=pd.get_dummies(x['Geography'],drop_first=True)
gender=pd.get_dummies(x['Gender'],drop_first=True)

x=pd.concat([x,states,gender],axis=1)

#drop old categorical variables
x=x.drop(['Geography','Gender'],axis=1)
```

```python
x
```

```python
# Now we split the data set into training and testing

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Now we are able to start making the ANN

import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(activation="relu", input_dim=11, units=6, kernel_initializer="uniform"))

# Adding the second hidden layer
classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))

# Adding the output layer
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 50)
```

```python
# Now that the ANN is built we can make predictions

y_pred = classifier.predict(X_test)
#If its bigger than 0.5 give me 1, other wise 0.
y_pred = (y_pred > 0.5)

# Confusion Matrix
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy=accuracy_score(y_test,y_pred)
```
